```rust
extern crate plotters;
use plotters::prelude::*;
use plotters::coord::Shift;

pub fn sierpinski_carpet(
    depth: u32, 
    drawing_area: &DrawingArea<SVGBackend, Shift>) 
-> Result<(), Box<dyn std::error::Error>> {
    if depth > 0 {
        let sub_areas = drawing_area.split_evenly((4,4));
        for (idx, sub_area) in (0..).zip(sub_areas.iter()) {
            if idx == depth {
                sub_area.fill(&WHITE)?;
            } else {
                sierpinski_carpet(depth - 1, sub_area)?;
            }
        }
    }
    Ok(())
}
evcxr_figure((4800,4800), |root| {
    root.fill(&BLACK)?;
    sierpinski_carpet(4, &root)
}).style("width: 400px")  /* You can add CSS style to the result */
```

```rust
let width = 400;
let height = 400;

pub fn bit_field(
    w: i32,
    h: i32,
    drawing_area: &DrawingArea<SVGBackend, Shift>
) -> Result<(), Box<dyn std::error::Error>> {
    for x in 1..w {
        for y in 1..h {
            match (x ^ y) % 17 {
                0 => {drawing_area.draw_pixel((x,y), &WHITE);()},
                1 => {drawing_area.draw_pixel((x,y), &RED);()}
                4 => {drawing_area.draw_pixel((x,y), &YELLOW);()},
                _ => (),
            };
        }
    }
    Ok(())
}
    
pub fn hsl_bit_field(
    w: i32,
    h: i32,
    drawing_area: &DrawingArea<SVGBackend, Shift>
) -> Result<(), Box<dyn std::error::Error>> {
    for x in 1..w {
        for y in 1..h {
            let s = (((x * 64) % y) as f64) / 40.;
            // println!("{}",s);
            drawing_area.draw_pixel((x,y), &HSLColor(s, 0.5, 0.5));
        }
    }
    Ok(())
}

evcxr_figure((width, height), |root| {
    root.fill(&BLACK)?;
    bit_field(width as i32, height as i32, &root);
    // hsl_bit_field(width as i32, height as i32, &root)
    Ok(())
}).style(format!("width: {}px", width))
```
